```python
import numpy as np
import scipy
import re
from scipy.spatial import distance

file_name = 'Sentences.txt'
with open(file_name) as f:
    lines = f.readlines()
sentences=[]
text= []
for line in lines:
    line = line.lower()
    line = line.strip()
    sentence = re.split('[^a-z]', line)
    for i in sentence:
        if i not in text:
            text.append(i)
    sentences.append(sentence)

while '' in text:
    text.remove('')
    
matrix = np.zeros((len(sentences), len(text)), dtype = int)
count = 0
for i in text:
    for j in range(len(sentences)):
        count = sentences[j].count(i)
        matrix[j][text.index(i)]=count
print(matrix.shape)
cosine =[]
for i in range(1, 22):
    cosine.append(distance.cosine(matrix[0], matrix[i]))

sorted_cosine = sorted(cosine)
print(cosine)
print(sorted_cosine)
with open('submission-1.txt', 'w') as f_w:
    f_w.write(str((cosine.index(sorted_cosine[1])+1))+' '+ str((cosine.index(sorted_cosine[0])+1)))

```
